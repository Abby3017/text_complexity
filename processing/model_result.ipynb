```python

import pickle
import matplotlib.pyplot as plt
import numpy as np
import sys
import os
```

```python
gru_5_run = [0.844, 0.844, 0.844]
gru_25_run = [0.886, 0.886, 0.886]
gru_50_run = [0.899, 0.899, 0.898]
gru_75_run = [0.906, 0.904, 0.905]
gru_100_run = [0.909, 0.909, 0.909]
```

```python
roberta_5_run = [0.909, 0.905,0.897]
roberta_25_run = [0.916, 0.922, 0.924]
roberta_50_run = [0.925, 0.926, 0.923]
roberta_75_run = [0.926, 0.927,0.925]
roberta_100_run = [0.928, 0.928, 0.928]
```

```python


# list of all data
data_gru = [gru_5_run, gru_25_run, gru_50_run, gru_75_run, gru_100_run]
data_roberta = [roberta_5_run, roberta_25_run, roberta_50_run, roberta_75_run, roberta_100_run]

# calculate averages
averages_gru = [np.mean(run) for run in data_gru]
averages_roberta = [np.mean(run) for run in data_roberta]

# create scatter plot
for i, run in enumerate(data_gru):
    jitter = np.random.normal(0, 0.1, size=len(run))
    run_roberta = data_roberta[i]
    if i == 4:
        plt.scatter([i+1] * len(run), run, color='blue', s=5)
        plt.scatter([i+1 + jitter], run_roberta, color='green', s=5)
    else:
        plt.scatter([i+1 + jitter], run, color='blue', s=5)
        plt.scatter([i+1 + jitter], run_roberta, color='green', s=5)

# create line plot for averages
plt.plot(range(1, len(data_gru)+1), averages_gru, color='red')
plt.plot(range(1, len(data_roberta)+1), averages_roberta, color='black')

# set xticks and labels
plt.xticks(range(1, len(data_gru)+1), ['5', '25', '50', '75', '100'])
plt.xlabel('% of training examples used')
plt.ylabel('Pearson Coefficient')

# set legend
plt.legend(['GRU', 'RoBERTa'])



plt.savefig('./../result/pearson_coeff_models_with_various_run.pdf', dpi=300, bbox_inches='tight')
```
