```python


# list of all data
data = [gru_5_run, gru_25_run, gru_50_run, gru_75_run, gru_100_run]

# calculate averages
averages = [np.mean(run) for run in data]

# create scatter plot
for i, run in enumerate(data):
    jitter = np.random.normal(0, 0.1, size=len(run))
    if i == 4:
        plt.scatter([i+1] * len(run), run, color='blue', s=5)
    else:
        plt.scatter([i+1 + jitter], run, color='blue', s=5)

# create line plot for averages
plt.plot(range(1, len(data)+1), averages, color='red')

# set xticks and labels
plt.xticks(range(1, len(data)+1), ['5', '25', '50', '75', '100'])
plt.xlabel('% of training examples used')
plt.ylabel('Validation loss')


plt.show()
```
